```python
col = ['Symptom Severity (1-10)',  'Mood Score (1-10)', 'Sleep Quality (1-10)', 'Physical Activity (hrs/week)' , 'Treatment Duration (weeks)', 'Stress Level (1-10)', 'Treatment Progress (1-10)', 'Adherence to Treatment (%)']

fig, axes = plt.subplots(nrows = 3, ncols = 3, figsize = (15, 10))
axes = axes.flatten()

for i, col in enumerate(col):
    sns.boxplot(y=df['Age'], ax=axes[i])
    axes[i].set_title(f'{col} Boxplot')

plt.tight_layout()
plt.show()
```
